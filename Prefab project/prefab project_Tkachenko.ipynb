### Корреляция между отзывами и продажами

```python
def relation_sum_score(platform):# функция для расчета корреляции и построения диаграммы рассеивания
    p = good_data_actual[good_data_actual['platform']==platform]
    print(p['user_score'].corr(good_data['sum_sales']).round(2))
    print(p['critic_score'].corr(good_data['sum_sales']).round(2))
    fig, (axis1,axis2) = plt.subplots(1,2, figsize=(15,6))
    p.plot(ax=axis1,kind='scatter', y='sum_sales', x='user_score', grid=True, alpha=0.6,\
           title = 'Зависимость между продажами и оценками пользователей')
    p.plot(ax=axis2,kind='scatter', y='sum_sales', x='critic_score', grid=True, alpha=0.6, \
           title = 'Зависимость между продажами и оценками критиков')
```

Рассмотрим как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков

```python
relation_sum_score('PS4')
```

Корреляция между оценкой пользователей и продажами очень слабая. Между оценкой критиков и продажами-слабая

## Проверка гипотез

### Гипотеза по платформам

Нулевая и альтернативные гипотезы:

Hо: средние пользовательские рейтинги платформ Xbox One и PC равны

H1: средние пользовательские рейтинги платформ Xbox One и PC не равны

```python
# срезы по платформам
xone = good_data_actual.loc[(good_data_actual['platform']== 'XOne') & (good_data_actual['user_score']\
                                                                       .notna()), 'user_score']
pc = good_data_actual.loc[(good_data_actual['platform']== 'PC') & (good_data_actual['user_score']\
                                                                   .notna()), 'user_score']
```

```python
results = st.ttest_ind(xone,pc, equal_var=False)#  дисперсии выборок не равны
alpha = .05 # критический уровень статистической значимости - общепринятое значение
# если p-value окажется меньше него - отвергнем гипотезу

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Согласно расчетам отвергнуть гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC равны нет оснований.

### Гипотеза по жанрам

Нулевая и альтернативные гипотезы:

Hо: средние пользовательские рейтинги жанров Action и Sports равны

H1: средние пользовательские рейтинги жанров Action и Sports не равны

```python
# срезы по жанрам
action = good_data_actual.loc[(good_data_actual['genre']== 'Action') & (good_data_actual['user_score']\
                                                                        .notna()), 'user_score']
sport = good_data_actual.loc[(good_data_actual['genre']== 'Sports') & (good_data_actual['user_score']\
                                                                       .notna()), 'user_score']
```

```python
results = st.ttest_ind(action,sport, equal_var=False)#  дисперсии выборок не равны
alpha = .05 # критический уровень статистической значимости- общепринятое значение
# если p-value окажется меньше него - отвергнем гипотезу

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Согласно уровню статистической значимости, есть основания отвергнуть гипотезу, что средние пользовательские рейтинги жанров Action и Sports равны
